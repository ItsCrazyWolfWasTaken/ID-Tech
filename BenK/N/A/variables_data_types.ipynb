**Variables** are like boxes; they hold important information, or data. A variable can hold different data types, such as numbers or words. When you're writing the name of a variable in Python, it should always be lowercase, with words separated by an underscore.

# Data Types
In this section, you'll explore Python's three commonly used data types: **integers**, **booleans**, and **strings**.

## **Integers**
One type of data or information a variable can hold is an **integer**. An integer is represented by whole numbers such as 0, 9, -100, 23, -77, etc.

In the example above, you already declared an integer variable by assigning 3 to the **number_of_doughnuts** variable. In the following example, you'll change the number of doughnuts in the chest.

```python
# Declare an integer variable
num_doughnuts = 3
print(num_doughnuts)
```

## **Boolean**
Another type of data a variable can hold is a **boolean**. A boolean determines whether a condition in a program is true or false.

In the following example, you'll create a variable to check if the doughnuts exist.

```python
# Declare a boolean variable
doughnuts_there = True
print(doughnuts_there)
```

## **Strings**
A **string** is a series of characters, words, or digits surrounded by double quotation marks. An empty string is also represented by double quotation marks.

In the previous lesson, you created strings when printing statements. In the following example, you'll name the doughnuts stored in the chest.

```python
# Declare a string variable
dougnut_type = ("Orange")
print(doughnut_type)
```
